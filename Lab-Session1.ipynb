```python
# !pip install sklearn
# !pip install scipy
import sklearn
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = df.dropna(how = 'any')
```

```python
my_x = df[['Age']]  #df[['Age','Pclass', 'Parch', 'Fare']]
my_x.head()
```

```python
my_y = df[["Survived"]]
my_y.sample(4)
```

```python
X_train, X_test, y_train, y_test = train_test_split(my_x, my_y, test_size=0.2)
```

```python
# Create linear regression object
first_regr = linear_model.LinearRegression()
```

```python

# Train the model using the training sets
first_regr.fit(X_train, y_train)
```

```python

# The coefficients
print('Coefficients: \n', first_regr.coef_)
```

```python
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((first_regr.predict(X_test) - y_test) ** 2))
```

```python
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % first_regr.score(X_test, y_test))
```

```python

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, first_regr.predict(X_test), color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
